# Lecture 1: Introduction to MHRQI
## Multi-scale Hierarchical Representation of Quantum Images

Welcome to the first lecture on **MHRQI**. In this series, we will explore a modern framework for representing images on quantum computers using a hierarchical approach. 

---

### 1. What is MHRQI?

MHRQI stands for **Multi-scale Hierarchical Representation of Quantum Images**. 

Traditional Quantum Image Representations (QIR) like **FRQI** (Flexible Representation of Quantum Images) and **NEQR** (Novel Enhanced Quantum Representation) typically map pixel coordinates directly to qubit states. While effective, they often lack robustness against noise and don't exploit the natural spatial hierarchy found in images.

**MHRQI** changes this by using a **Quad-tree decomposition** to represent images. Instead of seeing an image as a flat grid of pixels, MHRQI sees it as a nested structure of sub-regions.

### 2. The Core Concept: Hierarchy

Imagine a 16x16 image. 
1. At the **top level (Level 0)**, we have the whole image.
2. At **Level 1**, we divide it into 4 quadrants (Top-Left, Top-Right, Bottom-Left, Bottom-Right).
3. At **Level 2**, we divide each of those quadrants into 4 smaller sub-quadrants.
4. This continues until we reach the pixel level.

This hierarchy is mapped to **Hierarchical Coordinate Vectors (HCV)**.

### 3. Visualizing Quad-tree Decomposition

Let's look at how a pixel's location is represented in MHRQI.

### 4. Why Use a Hierarchy?

By encoding images hierarchically, MHRQI provides:
1. **Robustness**: Noise affecting higher-level qubits has a different "scale" of impact than lower-level qubits.
2. **Denoising Capabilities**: Sibling sub-regions in the quad-tree often share similar statistical properties. MHRQI exploits this for quantum denoising (which we will see in Lecture 3).
3. **Efficiency**: Multi-scale representation allows for adaptive encoding where smooth regions can be represented with fewer resources.

### 5. Summary

In MHRQI, an image is not just a collection of pixels; it's a **multi-scale quantum state**.

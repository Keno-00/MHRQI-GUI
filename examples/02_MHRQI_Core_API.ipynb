# Lecture 2: MHRQI Core API
## Hands-on with the MHRQI Library

In this lecture, we will learn how to use the `mhrqi` library to encode an image into a quantum circuit and simulate its reconstruction.

---

### 1. Preparing the Environment

First, we'll need a sample image. We use a small 16x16 image for quick simulation.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from mhrqi import MHRQI
from mhrqi.utils import general as utils

# Create a simple synthetic image (a white square on black background)
size = 16
img = np.zeros((size, size))
img[4:12, 4:12] = 1.0

plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.show()
```

### 2. Initializing MHRQI

The `MHRQI` class is the heart of the framework. It handles the creation of the quantum circuit and the mapping of registers.

To represent a $16 \times 16$ image, we need a depth of $\log_2(16) = 4$.

```python
depth = int(np.log2(size))
model = MHRQI(depth=depth)

print(f"Quantum Circuit initialized with {model.circuit.num_qubits} qubits.")
```

### 3. Data Uploading

We need to generate the **Hierarchical Coordinate Matrix** first, which maps coordinates to the quad-tree structure. Then we use `upload` to encode the image.

> **Tip**: For faster simulations, you can use `lazy_upload` which initializes the statevector directly.

```python
h_matrix = utils.generate_hierarchical_coord_matrix(size, d=2)

# Gate-based upload
model.upload(h_matrix, img)

print("Image data uploaded to quantum circuit.")
```

### 4. Simulation and Reconstruction

Running `simulate()` returns an `MHRQIResult` object. This object provides high-level methods to analyze the output.

```python
# Simulate with 10,000 shots
result = model.simulate(shots=10000)

# Reconstruct the image
recon = result.reconstruct()

# Plot the result
result.plot(title="Quantum Reconstruction")
```

### 5. Quality Metrics

MHRQI has built-in support for calculating image quality metrics like **MSE** and **SSIM**.

```python
metrics = result.compute_metrics(reference_image=img)
for name, val in metrics.items():
    print(f"{name}: {val:.4f}")
```
